```python
import json
import pandas as pd
import re
with open('C:/Users/sangho/Desktop/tqa_sample.json',encoding="utf-8") as f:
    jsonObj = json.load(f)
def flatten_json(y):
    out = {}
    def flatten(x, name=''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name + a + '_')
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, name + str(i) + '_')
                i += 1
        else:
            out[name[:-1]] = x
    flatten(y)
    return out
flat = flatten_json(jsonObj)
results = pd.DataFrame()
special_cols = []
columns_list = list(flat.keys())
for item in columns_list:
    try:
        row_idx = re.findall(r'\_(\d+)\_', item )[0]
    except:
        special_cols.append(item)
        continue
    column = re.findall(r'\_\d+\_(.*)', item )[0]
    column = column.replace('_', '')
    row_idx = int(row_idx)
    value = flat[item]
    results.loc[row_idx, column] = value
for item in special_cols:
    results[item] = flat[item]
results.to_csv('C:/Users/sangho/Desktop/file.csv', index=False)
```
